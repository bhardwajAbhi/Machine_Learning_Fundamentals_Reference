# Cleaning Data

```python
# Cleaning up the data
clean1 = data[data['Hour'] < 6]
clean2 = data[data['Hour'] > 18]

print("Cleaning Data from 12:00 am to 5:59 am")
clean1.head(10)
```

```python
print("Cleaning Data from 18:00 pm to  12.00 am")
clean2.head(10)
```

```python
data =  data.drop(clean1.index, axis=0)
data =  data.drop(clean2.index, axis=0)
```

## Input and Ouput Data

```python
x = data.drop(['Solar Radiation (GHI)'], axis=1)
y = data.iloc[:, 10].values 
```

```python
from sklearn.preprocessing import MinMaxScaler
x = pd.DataFrame(MinMaxScaler().fit(x).transform(x))
```

```python
col = ['Year', 'Month', 'Day', 'Hour', 'Dew Point', 'Temperature', 'Pressure',
       'Relative Humidity', 'Wind Direction', 'Wind Speed',]
```

# <font color = "green"> SelectKBest Method </font>

- **Using Chi Square Test**
- **Extracting Top 4 features**

```python
bestfeatures = SelectKBest(score_func = chi2, k = 10)
fit = bestfeatures.fit(x, y)
```

```python
scores = pd.DataFrame(fit.scores_)
column = pd.DataFrame(col)
```

```python
# contatinating data_features with the scores
featureScores = pd.concat([column, scores], axis=1)

#naming the dataframe columns
featureScores.columns = ['Features', 'Score'] 
```

```python
#print 4 best features
print(featureScores.nlargest(10, 'Score'))  
```

## Visualisation

```python
featureScores.nlargest(10, 'Score').plot(kind='bar')
plt.show()
```
