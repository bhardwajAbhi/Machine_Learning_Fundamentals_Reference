```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Loading Data

```python
data = pd.read_csv("C:\\Users\\dell\\Desktop\\bank_note_data.csv")
```

```python
data.head()
```

```python
data['Class'].unique()
```

```python
data.info()
```

```python
features = ['Image.Var', 'Image.Skew', 'Image.Curt', 'Entropy']
targets  = ['Class']
```

## Scaling Data

```python
data_norm = data.copy()
 
sc = StandardScaler()
data_norm[features] = sc.fit(data[features]).transform(data[features])
```

# <font color = "green">TSNE </font>

```python
tsne = TSNE(learning_rate = 500, n_components = 2)

X_tsne = tsne.fit_transform(data_norm[features])
```

```python
y = data[targets]
```

### Visualisation

```python
plt.figure(figsize=(16, 11))
plt.scatter(X_tsne[np.where(y == 0), 0], X_tsne[np.where(y == 0), 1], color='g', marker='*')
plt.scatter(X_tsne[np.where(y == 1), 0], X_tsne[np.where(y == 1), 1], color='r', marker='.')

plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('T-SNE')
plt.show()  
```
