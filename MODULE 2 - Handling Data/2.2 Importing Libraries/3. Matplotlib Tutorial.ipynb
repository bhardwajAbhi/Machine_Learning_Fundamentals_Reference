## Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line Plot

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.plot(x, y)

# Formatting the line type  
plt.plot(x,y,'^') 

plt.title("Line Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

```python
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x)

plt.plot(x, y)
plt.title("Line Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

### Scatter Plot

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

### Comparison

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.plot(x, y, color = 'g', linewidth = 6, label = 'first')
plt.plot(x2, y2, color = 'r', linewidth = 4, label = 'second')

plt.title("Comparison Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding Notation
plt.annotate(xy=[2, 3], s='First Line') 
plt.annotate(xy=[4,4], s='Second Line') 

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=4)

plt.show
```

## Vertical Bar Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.bar(x, y, color = 'g', linewidth = 6, label = 'first')
plt.bar(x2, y2, color = 'r', linewidth = 4, label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=2)
plt.show
```

## Horizontal Bar Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.barh(x, y, color = 'g', linewidth = 6, label = 'first')
plt.barh(x2, y2, color = 'r', linewidth = 4, label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=2)
plt.show
```

### Bubble Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.scatter(x, y, s=np.random.rand(5)*1000, c =np.random.rand(5))
plt.title("Scater Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

## Pie Chart

```python
label = ['a', 'b', 'c', 'd', 'e']

# Exploding 2nd slice
explode = (0, 0.1, 0, 0, 0)

data = [3 ,7, 4, 12 ,5]
plt.pie(data, explode=explode, labels = label, shadow = True)
plt.show()
```
